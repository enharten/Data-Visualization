```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
unrate = pd.read_csv('UNRATE.csv')
```

```python
bunrate = pd.read_csv('LNS14000006.csv')
```

```python
nairu = pd.read_csv('NROUST.csv')
```

```python
df = pd.merge(unrate,bunrate,how='inner')
df.columns = ['Date','Unemployment Rate','Black Unemployment Rate']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
nairu['DATE'] = pd.to_datetime(nairu['DATE'])
nairu = nairu.rename(columns={'DATE':'Date'})
```

```python
fig, ax = plt.subplots(figsize=(30,12))
plt.rcParams['lines.linewidth'] = 4
ax.set_xlim([datetime.date(1970, 1, 1), datetime.date(2025, 1, 1)])
ax.plot('Date','Unemployment Rate',data=df)
ax.plot('Date','Black Unemployment Rate',data=df)
ax.plot('Date','NROUST',data=nairu)
ax.set(xlabel='Date', ylabel='Percent',
       title='Unemployment Rate')
ax.legend(frameon=False)

plt.show()
```
